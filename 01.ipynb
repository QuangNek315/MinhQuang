```python
# Thêm thư viện
from PIL import Image
```

```python
# Khai báo một biến chứa thư mục
dir = 'D:/DangDucMinhQuang/AnhDep'
```

```python
# Khai báo đường dẫn ảnh
image_path = dir + "/01.jpg"
print(image_path)
```

```python
# Sử dụng Image.open() để đọc ảnh
AnhDep01 = Image.open(image_path)
```

```python
# Hiển thị ảnh
display(AnhDep01)
```

```python
# In thông tin cơ bản
print("Định dạng ảnh: " + AnhDep01.format)
```

```python
#In kích cỡ ảnh
print("Kích cỡ ảnh: ", AnhDep01.size)
```

```python
#Lưu ảnh và chuyển đổi ảnh
new_AnhDep_01_path = dir + "/new_01.jpg"
new_AnhDep_png_01_path = dir + "/new_01.png"
AnhDep01.save(new_AnhDep_01_path)
AnhDep01.save(new_AnhDep_png_01_path)
```

```python
# Đóng tập tin
AnhDep01.close()
```
