```python
# Định vị khu vực cắt
region = (100,75, 175, 150)
# Cắt vùng ảnh
cropped_image = AnhDep.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị cần dán
paste_position = (600, 200)
AnhDep.paste(cropped_image, paste_position)
display(AnhDep)
```
